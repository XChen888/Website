```python
# Find out wrong column "Date" from bixler
bixler = pd.read_csv("C:/Users/chenx/Desktop/SA/04_29/BixLer-2022-04-29.csv")
bixler_filter = bixler[bixler["Date"] != "2022-04-29"]
bixler_filter
```

```python
# Chnange UTC-time to local time, generate a new column "Time_new",and put "Time_new" into bixler

from datetime import datetime
FMT = '%H:%M:%S.%f'
time = bixler["Time"]
time_null = (bixler.iloc[0]["Time"])
print("The first UTC time: " + time_null)

time_us = []
for key,value in bixler["Time"].iteritems():
    time_def = datetime.strptime(value, FMT) - datetime.strptime(time_null, FMT) 
    time_us_ref = time_def.total_seconds() * 1000000 
    #print(key, time_us_ref)
    time_us.append(time_us_ref)
#print(time_us)
bixler["time_us"] = time_us
  
# Rerange the position of column "Time_new_us"

titles = list(bixler.columns)
print(titles)

bixler = bixler[['time_us', 'Date', 'Time', 'RSSI(dB)', 'RxBt(V)', 'Cels(V)', 'GPS', 'GAlt(m)', 'GSpd(kts)', 'Date.1', 'Rud', 'Ele', 'Thr', 'Ail', 'S1', 'S2', 'SA', 'SB', 'SC', 'SD', 'SF', 'SH', 'LSW', 'TxBat(V)']]
#bixler.to_csv("C:/Users/chenx/Desktop/SA/04_29/BixLer_Test.csv", index = False)
bixler
```

```python
plt.plot(bixler["Time_new_us"], bixler["Ail"])
plt.title("Aileron deflection measurement of Bixler Aileron")
plt.xlabel("Time [us]")
plt.ylabel("Ail [??]") 
#plt.xticks(np.arange(0, 5.5 * 10 ** 8, 0.5 * 10 ** 8))
#plt.legend()
plt.grid()
plt.show()
```
